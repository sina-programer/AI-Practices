```python
import numpy as np
import seaborn as sns
from scipy import stats
from matplotlib import pyplot as plt
```

```python
data_normal = stats.norm.rvs(loc=0, scale=1, size=5000)
# data_normal = np.random.normal(loc=0, scale=1, size=5000)
print(data_normal)
sns.histplot(data_normal)
```

```python
data_uniform = stats.uniform.rvs(loc=0, scale=10, size=5000)
# data_uniform = np.random.uniform(low=0, high=10, size=5000)
print(data_uniform)
sns.histplot(data_uniform)
```

```python
data_bernoulli = stats.bernoulli.rvs(p=.2, size=5000)
# data_bernoulli = np.random.binomial(n=1, p=.2, size=5000)
print(data_bernoulli[:100])
sns.histplot(data_bernoulli)
```

```python
data_binom = stats.binom.rvs(n=100, p=.2, size=5000)
# data_ninom = np.random.binomial(n=100, p=.2, size=5000)
print(data_binom[:100])
sns.histplot(data_binom)
```

```python
data_multinomial = stats.multinomial.rvs(n=100, p=[.2, .3, .5], size=5000)
# data_multinomial = np.random.multinomial(n=100, pvals=[.2, .3, .5], size=5000)
print(data_multinomial[:15])
```

```python
data_poisson = stats.poisson.rvs(mu=5, size=5000)
# data_poisson = np.random.poisson(lam=5, size=5000)
print(data_poisson)
sns.histplot(data_poisson)
```

```python
data_exponential = stats.expon.rvs(scale=3, size=5000)
# data_exponential = np.random.exponential(scale=3, size=5000)
print(data_exponential)
sns.histplot(data_exponential)
```
